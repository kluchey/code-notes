## Task 3:  What type of pattern is this?

In this next task, you will identify what type of pattern is being generated by using vectors and something called a *parameter*, given by the value of `t`.

You will also be introduced to a coding concept called *loops* which allow us to repeat a set of instructions a given number of times. Note: **Python always starts its counting at 0, so the list 0, 1, 2, 3 is a list of the first 4 numbers**.

4. How was this pattern generated using vectors? Why does it make sense that this pattern is that type?

```python
P = np.array([1,1])  # Position Vector
d = np.array([2, 1]) # direction vector
table = []           # An empty list which will be loaded with vectors later

# Note! The table will be unused in this activity but this setup can be used by us in Unit 6.

print('''t   |   x   |    y
------------------''')

for t in range(4):   # This is a for-loop, run 4 times since range(4) is 0, 1, 2, 3.
    v = P + t*d
    table.append(v)  # This adds the vector v to our table from earlier.
    print(f'{t}   |   {v[0]}   |    {v[1]}')
    
```
